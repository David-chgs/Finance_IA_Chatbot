```python
import os
from groq import Groq

os.environ["GROQ_API_KEY"] = "sua_chave_aqui"

client = Groq()
```

```python
SYSTEM_PROMPT = """
Você é um especialista em finanças pessoais e investimentos.

Responda sempre:
- De forma simples e didática
- Com exemplos práticos
- Em no máximo 5 linhas
- Evitando termos técnicos complexos

Se fizer sentido, use analogias do dia a dia.
"""
```

```python
messages = [
    {"role": "system", "content": SYSTEM_PROMPT}
]
```

```python
def ask_finance_bot(user_input):
    messages.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=messages,
        temperature=0.5
    )

    answer = response.choices[0].message.content

    messages.append({"role": "assistant", "content": answer})

    return answer
```

```python
def run_chat():
    print("💰 Assistente Financeiro iniciado! (digite 'sair' para encerrar)\n")

    while True:
        user_input = input("Você: ")

        if user_input.lower() in ["sair", "exit", "quit"]:
            print("Encerrando...")
            break

        response = ask_finance_bot(user_input)
        print(f"\n🤖 Assistente: {response}\n")
```

```python
run_chat()
```
